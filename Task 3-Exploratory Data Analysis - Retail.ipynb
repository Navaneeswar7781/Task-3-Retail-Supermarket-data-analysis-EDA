we are talking about geographycal location then, from 'Washington' state there are more selling products, after that 'New York' and 'Texas'.

```python
state_sales = data.groupby('State').Sales.agg(['sum','mean','min','max','count','median','std','var']).sort_values(by='mean',ascending=False).reset_index()
state_sales
```

```python
plt.figure(figsize=(20,20))

sns.barplot(x='State',y='mean',data=state_sales)
plt.xticks(fontsize=15)
plt.yticks(fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('State',fontsize=30)
plt.ylabel('Mean Sales',fontsize=30)
plt.title('State wise Mean Sales',fontsize=40)
plt.show()
```

Wyoming state have highest amount of sales from other states in US.
- South Dakota have lowest amount of sales as compare to other states.

```python
state_Discount = data.groupby('State').Discount.agg(['sum','mean','min','max','count','median','std','var']).sort_values(by='mean',ascending=False).reset_index()
state_Discount
```

```python
plt.figure(figsize=(20,17))

sns.barplot(x='State',y='mean',data=state_Discount)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xticks(rotation=90)
plt.xlabel('State',fontsize=30)
plt.ylabel('Mean Discount',fontsize=30)
plt.title('State wise Mean Discount',fontsize=40)
plt.show()
```
